```python
#Project 1:
#Profit prediction using LinearRegression
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
df=pd.read_csv("/content/Ice Cream Sales - temperatures.csv")
t=np.array([i for i in df["Temperature"]]).reshape(-1,1)
```

```python
p=df["IceCreamProfits"]
LR.fit(t,p)
```

```python
n=int(input("Enter Temperature:"))
profit=(LR.predict([[n]]))
print(profit)

if(profit<-5 and profit>-10):
  print("Sales are very low, decrease production")
elif(profit>-5 and profit<10):
  print("Sales are linear please spend some more on marketing")
elif(profit>10):
  print("Sales are rapid, increase production")
else:
  print("Margins are negative, take a break from business :)")
```

```python
from matplotlib import pyplot as plt
plt.scatter(t,p,color='black')
plt.title('Linear Regression')
plt.show()
```
