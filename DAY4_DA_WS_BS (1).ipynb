```python
from bs4 import BeautifulSoup
```

```python
url ="https://www.w3schools.com/python/default.asp"
```

```python
title =code.find("title") #to find a specific part
print(title)
```

```python
print(title.text)#text without tags
```

EXTRACTING TABLE FROM A WEBPAGE
1.Import libraries
2.Save the url
3.Using requests.get() , acess the web page
4.Using BeautifulSoup(),access the HTML code

```python
url2= "https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1#:~:text=A%20record%20number%20of%20Indians%2C%20169%20in%20total%2C,eminent%20personalities%20like%20Gautam%20Adani%20and%20Uday%20Kotak."
```

```python
page2 = requests.get(url2)
```

```python
code2 = BeautifulSoup(page2.text)
```

```python
print(code2)
```

```python
table =code2.find("table") #to find a specific part
print(table)
```

```python
th=table.find_all('th')#to find more than one tags present in webpage
print(th)
```

```python
headings=[i.text for i in th]#using list comprehension, we have taken only the text part (.text) of th
print(headings)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=headings)#Creating a dataframe with the content we have.
print(mydf)
```

```python
rows=table.find_all('tr')#finding all rows of table
print(rows)
```

```python
data=table.find_all('td')
print(data)
```

```python
data_act=[i.text for i in data]
```

```python
print(data_act)
```

```python
len(data_act)
```

```python
ind=0#referring indexes
start=0
stop=4
for i in data_act:#elements in dataset
  if stop <= len(data_act):
    mydf.loc[ind]=data_act[start:stop]#cols already added before in dataframe

    start=start+4
    stop=stop+4
    ind=ind+1
mydf#dataframe created by placing rows

```

```python
mydf.to_csv("mydf2.csv")
```
