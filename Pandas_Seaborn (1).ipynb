```python
tips=sns.load_dataset("tips")
```

```python
#Showing the difference by differentiating colors - hue
sns.relplot(data=tips,x='total_bill',y="tip",hue="smoker")
```

```python
sns.relplot(data=tips,x='total_bill',y="tip",hue="sex")
```

```python
sns.relplot(data=tips,x='total_bill',y="tip",hue="smoker",style="time",palette="dark")#additional feature added to differentiate i.e style for time
```

```python
sns.relplot(data=tips,x='total_bill',y="tip",hue="size")
```
