```python
import pandas as pd
df=pd.read_csv("/content/demodt.txt",sep=',')
print(df)
```

```python
from sklearn.tree import DecisionTreeClassifier

features_cols=['Literacy','Cleanliness','Crime_Rate']
features=df[features_cols]#features
Target=df.Good#Target variable

model=DecisionTreeClassifier()
model.fit(features,Target)
x=int(input("enter Literacy rate: "))
y=int(input("enter Cleanliness: "))
z=int(input("enter Crime rate:"))
p=model.predict([[x,y,z]])#crime_rate>50-->Bad else good

if(p==[1]):
  print("Good")
else:
  print("Bad")


```

```python
for i,tree in enumerate(RF.estimators_):
  tree_dot_file = os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file = os.path.join(output_dir,f"tree_{i}.png")
```

```python
#dot file creates inside directory
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["Literacy", "Cleanliness", "Crime_Rate"],
                    class_names=[str(cls) for cls in RF.classes_], filled=True, rounded=True)
```

```python
command = f"dot -Tpng {tree_dot_file} -o {tree_png_file}"#spaces are sensitive
os.system(command)
print(f"Tree {i} visualization saved to {tree_png_file}")
```
